## Import data : 

```python
try : 
  data = open('/content/drive/MyDrive/POS_data.txt','r')
except : 
  print("File not found!!")

data = data.readlines()
```

```python
## remove all junk characters 
import re

for i in range(len(data)) :
  data[i] = data[i].strip()

for text in data :
  if text == '' or text.count('[') > 0 or text.count(']') > 0 or text.count('*') > 0  or text.count('-') or text.count('_'):
    data.remove(text)


data = [s for s in data if not re.search('\d.*[a-zA-Z]|[a-zA-Z].*\d', s)]


try : 
  file = open('/content/drive/MyDrive/new_POS_data.txt','w')
except : 
  print('Can\'t find file!!')

file.writelines(data)
```

```python
## use nltk

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk import pos_tag


nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

fullText = ""

for text in data : 
  fullText += text

## Tokenize the text : 

tokens =  word_tokenize(fullText)

## Remove stopwords :

filtered_words = [word for word in tokens if word.lower() not in stopwords.words('english')]

## Lemmatize : 

lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]


## Part of Speech tagging :

pos = pos_tag(lemmatized_words)

```
